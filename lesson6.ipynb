```python
import pandas as pd
import os

data_dir = "data/pascal_2007"

train_df = pd.read_csv(os.path.join(data_dir, "train.csv"))

train_df.head()
```

```python
from fastai.vision.all import DataBlock

dblock = DataBlock()
dsets = dblock.datasets(train_df)

len(dsets.train),len(dsets.valid)
```

```python
x, y = dsets.train[0]
x, y
```

```python
dblock = DataBlock(get_x=lambda r: r["fname"], get_y=lambda r: r["labels"])
dsets = dblock.datasets(train_df)

dsets.train[0]
```

```python
def get_x(r): return os.path.join(data_dir, "train", r["fname"])
def get_y(r): return r["labels"].split(" ")

dblock = DataBlock(get_x=get_x, get_y=get_y)
dsets = dblock.datasets(train_df)
dsets.train[0]
```

```python
from fastai.vision.all import ImageBlock, MultiCategoryBlock

dblock = DataBlock(blocks=[ImageBlock, MultiCategoryBlock],
                   get_x=get_x,
                   get_y=get_y)
dsets = dblock.datasets(train_df)
dsets.train[0]
```
